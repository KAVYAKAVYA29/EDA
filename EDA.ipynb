```python
# Import necessary libraries
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/PPGE.csv')

# Check the dimensions of the dataset
print("Dimensions:", data.shape)

# Display the first few rows to understand the structure
print("First few rows:\n", data.head())

# Display summary information
print("\nSummary Information:")
print(data.info())

# Display basic statistics for numeric columns
print("\nDescriptive Statistics:")
print(data.describe())
```

```python
# Check for missing values
print("\nMissing Values:\n", data.isnull().sum())

# Handle missing values (Example: Fill numeric NaNs with the mean)
data.fillna(data.mean(), inplace=True)

# Check for duplicates and remove them
duplicates = data.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
data.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plotting style
sns.set(style="whitegrid")

# Plot distributions for numerical features
for column in data.select_dtypes(include='number').columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()

# Plot count plots for categorical features
for column in data.select_dtypes(include='object').columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(y=data[column])
    plt.title(f'Count Plot of {column}')
    plt.show()
```

```python
# Pairwise relationships between numeric features
sns.pairplot(data.select_dtypes(include='number'))
plt.show()

# Correlation heatmap for numeric features
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()

# Box plots for numerical vs categorical features
for column in data.select_dtypes(include='object').columns:
    for num_col in data.select_dtypes(include='number').columns:
        plt.figure(figsize=(8, 4))
        sns.boxplot(x=column, y=num_col, data=data)
        plt.title(f'{num_col} by {column}')
        plt.show()
```
